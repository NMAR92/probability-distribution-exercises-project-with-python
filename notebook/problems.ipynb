# Probability Distribution Problems

**Binomial Distribution**

In probability theory and statistics, the binomial distribution is the discrete probability distribution that gives only two possible results in an experiment, for example, when you flip a coin it can by either Head or Tails.

```python
# Probability functions, density and distribution

def dbinom(x,size,prob=0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    from scipy.stats import binom
    result=binom.pmf(k=x,n=size,p=prob,loc=0)
    return result

def pbinom(q,size,prob=0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    from scipy.stats import binom
    result=binom.cdf(k=q,n=size,p=prob,loc=0)
    return result

def qbinom(p, size, prob=0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    from scipy.stats import binom
    result=binom.ppf(q=p,n=size,p=prob,loc=0)
    return result

def rbinom(n,size,prob=0.5):
    """
    Generates random variables from the binomial distribution
    """
    from scipy.stats import binom
    result=binom.rvs(n=size,p=prob,size=n)
    return result
```

```python
dbinom(2, size = 10, prob = 0.2) #Probability that a binomial (10,0.2) takes the value 2,
pbinom(2, size = 10, prob = 0.2) #Probability that a binomial (10,0.2) will take a value less than 2
qbinom(0.9, size = 10, prob = 0.2) # what value of a binomial (10,0.2) has a cumulative probability of 0.9?
rbinom(2000, size = 10, prob = 0.2) # generate 2000 random values from a binomial distribution (10,0.2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Discrete distributions
xs = np.arange( 1, 11 )
ys = dbinom(xs,size=10,prob=0.5) # compute the shape of the distribution
plt.plot( xs, ys, 'o' )   # plot circles...
plt.vlines( xs, 0, ys )   # ...and lines
plt.ylim( bottom=0 )      # ensure sensible bottom border
plt.title('Probability function binomial distribution')
plt.show()
```

```python
#Normal
# Probability functions, density and distribution

def dnorm(x, mean, sd):
    """
    Calculates the point estimate of the normal distribution
    """
    from scipy.stats import norm
    result=norm.pdf(x,loc=mean,scale=sd)
    return result

def pnorm(q, mean, sd):
    """
    Calculates the cumulative of the normal distribution
    """
    from scipy.stats import norm
    result=norm.cdf(q,loc=mean,scale=sd)
    return result

def qnorm(p, mean, sd):
    """
    Calculates the quantile function from the normal distribution
    """
    from scipy.stats import norm
    result=norm.ppf(p,loc=mean,scale=sd)
    return result

def rnorm(mean, sd, size):
    """
    Generates random variables from the Normal distribution
    """
    from scipy.stats import norm
    result=norm.rvs(loc=mean,scale=sd, size=size)
    return result

# Plot PDF
x = np.arange(-10, 10, 0.1)
plt.plot(x,dnorm(x,0,1))
plt.ylim( bottom=0 )      # ensure sensible bottom border
plt.title('Probability function Normal distribution PDF')
plt.show()

# Plot CDF
x = np.arange(-10, 10, 0.1)
plt.plot(x,pnorm(x,0,1))
plt.ylim( bottom=0 )      # ensure sensible bottom border
plt.title('Probability function Normal distribution CDF')
plt.show()

# Plot PPF
x = np.arange(0, 1, 0.01)
plt.plot(x,qnorm(x,0,1))
plt.title('Probability function Normal distribution PPF')
plt.show()

# Plot random numbers
np.random.seed(10)
y = rnorm(0, 1, 10000)
bin = np.arange(-4,4,0.1)  
plt.hist(y, bins=bin, edgecolor='blue') 
plt.show()
```

```python
#Uniform
# Probability functions, density and distribution

def duniform(x, start, end):
    """
    Calculates the point estimate of the uniform distribution
    """
    from scipy.stats import uniform
    result=uniform.pdf(x, loc=start, scale=end-start)
    return result

def puniform(z, start, end):
    """
    Calculates the cumulative of the uniform distribution
    """
    from scipy.stats import uniform
    result=uniform.cdf(z, loc=start, scale=end-start)
    return result

def quniform(q, start, end):
    """
    Calculates the quantile function from the uniform distribution
    """
    from scipy.stats import uniform
    result=uniform.ppf(q, loc=start, scale=end-start)
    return result

def runiform(start, end, size):
    """
    Generates random variables from the uniform distribution
    """
    from scipy.stats import uniform
    result=uniform.rvs(loc=start, scale=end-start, size=size)
    return result

# Plot PDF
x = np.arange(2,10,0.1)
plt.plot(x,duniform(x,3,8))
plt.ylim( bottom=0 )      # ensure sensible bottom border
plt.title('Probability function uniform distribution PDF')
plt.show()

# Plot CDF
x = np.arange(2,10,0.1)
plt.plot(x,puniform(x,3,8))
plt.ylim( bottom=0 )      # ensure sensible bottom border
plt.title('Probability function uniform distribution CDF')
plt.show()

# Plot PPF
x = np.arange(0, 1, 0.01)
plt.plot(x,quniform(x, 3, 8))
plt.title('Probability function uniform distribution PPF')
plt.show()

# Plot random numbers

np.random.seed(10)
y = runiform(3, 8, 10000)
bin = np.arange(0,10,0.1)  
plt.hist(y, bins=bin, edgecolor='blue') 
plt.show()
```
